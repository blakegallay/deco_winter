```python
import csv
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib as mpl

# sample to plot
# include all phi
energy = '1GeV'
thetas = [0.0, 25.8, 60.0, 78.5]

muon_dir = './simulation-test/simulation/sample_stats/mu+/'
track_file = './deco/Track.csv'
spot_file = './deco/Spot.csv'

#tracks = pd.read_csv(track_file)
#spots = pd.read_csv(spot_file)
```

```python
# we are plotting the distribution of classification probabilities for single samples
# i.e. set E, theta

muons = {}

for theta in thetas:
    muons[theta] = pd.DataFrame()
    for path in os.listdir(muon_dir):
        if energy in path and ('theta_'+str(theta)) in path:
            muons[theta] = muons[theta].append(pd.read_csv(muon_dir + path), ignore_index = True)
```

```python
for theta in thetas:

    p_spot = muons[theta]['p_spot'].tolist()

    p_track = muons[theta]['p_track'].tolist()

    p_worm = muons[theta]['p_worm'].tolist()

    fig = plt.figure(figsize=(10, 8))
    plt.ylabel("Count")
    #plt.yscale("log")
    plt.xlabel("Probability")
    plt.title("Distribution of Classification \n mu+, E="+ str(energy) + ", theta=" + str(theta) + "deg, all phi")

    plt.hist(p_spot, bins = 50, alpha=0.5, label='p_spot', range=(0,1))
    plt.hist(p_track, bins = 50, alpha=0.5, label='p_track', range=(0,1))
    plt.hist(p_worm, bins = 50, alpha=0.5, label='p_worm', range=(0,1))
    plt.legend(loc='upper right')

    #plt.savefig('./width_dist.png')

    plt.show()

    plt.close()

```
